```python
import pandas as pd

# Đọc dữ liệu từ tệp CSV
pizza_sales_df = pd.read_csv('/content/drive/MyDrive/pizza_sales.csv')

# In ra các thông tin cơ bản về dữ liệu
print("Thông tin cơ bản về dữ liệu:")
print(pizza_sales_df.info())

# Hiển thị một số dòng đầu tiên của DataFrame để xem dữ liệu trước khi làm sạch
print("\nDữ liệu trước khi làm sạch:")
print(pizza_sales_df.head())

# Làm sạch dữ liệu:
# 1. Kiểm tra và xử lý dữ liệu trùng lặp, dữ liệu thiếu, hoặc không hợp lệ nếu có.
# 2. Chuyển đổi kiểu dữ liệu của các cột nếu cần.
# 3. Loại bỏ các cột không cần thiết.

# Ví dụ:
# Loại bỏ cột 'pizza_id' nếu không cần thiết
pizza_sales_df.drop(columns=['pizza_id'], inplace=True)

# Kiểm tra và xử lý dữ liệu trùng lặp
duplicate_rows = pizza_sales_df.duplicated()
if duplicate_rows.any():
    print("\nCó dữ liệu trùng lặp. Đang loại bỏ...")
    pizza_sales_df.drop_duplicates(inplace=True)
    print("Dữ liệu sau khi loại bỏ dữ liệu trùng lặp:")
    print(pizza_sales_df.head())

# Kiểm tra và xử lý dữ liệu thiếu nếu có
missing_data = pizza_sales_df.isnull().sum()
if missing_data.any():
    print("\nDữ liệu có dữ liệu thiếu. Đang xử lý...")
    # Xử lý dữ liệu thiếu tại đây, ví dụ: điền giá trị trung bình hoặc loại bỏ dòng có dữ liệu thiếu
    # pizza_sales_df.fillna(value=0, inplace=True)
    print("Dữ liệu sau khi xử lý dữ liệu thiếu:")
    print(pizza_sales_df.head())

# Kiểm tra lại thông tin cơ bản sau khi làm sạch dữ liệu
print("\nThông tin cơ bản về dữ liệu sau khi làm sạch:")
print(pizza_sales_df.info())

# Lưu dữ liệu sau khi làm sạch vào một tệp mới nếu cần
# pizza_sales_df.to_csv('cleaned_pizza_sales.csv', index=False)
```
